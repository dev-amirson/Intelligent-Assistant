```python
import pandas as pd
```

```python
# reading csv file
Dataset=pd.read_csv("ROCStories__spring2016 - ROCStories_spring2016.csv")
```

```python
Dataset.drop(columns=["storyid"],axis=1,inplace=True)
```

```python
Dataset.head()
```

```python
Dataset["output"] = Dataset["sentence1"] + Dataset["sentence2"]++ Dataset["sentence3"]+ Dataset["sentence4"]+ Dataset["sentence5"]
```

```python
Dataset.head()
```

```python
Dataset.drop(columns=["sentence1","sentence2","sentence3","sentence4","sentence5",],axis=1,inplace=True)
```

```python

Dataset = Dataset.rename(columns={"storytitle":"source_text", "output":"target_text"})
Dataset = Dataset[['source_text', 'target_text']]
```

```python
Dataset.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(Dataset, test_size=0.2)
train_df.shape, test_df.shape
```

```python
!pip install simplet5
```

```python
from simplet5 import SimpleT5
```

```python
model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
model.train(train_df=train_df[:32000],
            eval_df=test_df[:1000], 
            source_max_token_len=50, 
            target_max_token_len=300, 
            batch_size=8, max_epochs=2, use_gpu=True)
```
